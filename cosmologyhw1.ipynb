```python
import math
import fileinput

import numpy as np
import scipy as sp
import scipy.interpolate

import matplotlib.pyplot as plt

from scipy.integrate import quad
from scipy.misc import derivative
from scipy.integrate import odeint

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})

rcParams.update({'figure.autolayout': True})
plt.rc('font',family='serif')
```

```python
def f(x,A,B):
    result = pow(A*pow(x,2.0) + B*pow(x,4.0),0.5)
    return result
```

```python
print('f(1,0.8,0.2)=',f(1,0.8,0.2))
print('f(2,0.8,0.2)=',f(2,0.8,0.2))
print('f(3,0.8,0.2)=',f(3,0.8,0.2))
print('f(100,0.8,0.2)=',f(100,0.8,0.2))
```

```python
integral = quad(f,0,100.,args=(0.8,0.2))
print('integral 0 to 100 = ',integral)
print('just integral = ',integral[0])
```

```python
x_list = np.linspace(0,100,101)

f_list = f(x_list,0.8,0.2)

print('f(100,0.8,0.2)=',f_list[x_list==100][0])
```

```python
integral_list = np.array([])

for x in x_list:
    integral = quad(f,0,x,args=(0.8,0.2))[0]
    integral_list = np.concatenate([integral_list,[integral]])
```

```python
#1a
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_xlabel('x',fontsize=20)
ax.set_ylabel('integral f(x)',fontsize=20)

ax.plot(x_list,integral_list,'b',linewidth=1.5,label='A = 0.8, B = 0.2')
ax.legend(loc='lower right')
```

```python
#1b
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_xlabel('integral f(x)',fontsize=20)
ax.set_ylabel('x',fontsize=20)

ax.plot(integral_list,x_list,'b',linewidth=1.5,label='A = 0.8, B = 0.2')
ax.legend(loc='lower right')
```
